## **Combined Tune Models - best hyperparameters**


Random Forest: Best Parameters: {'max_depth': 50, 'max_features': 4, 'min_samples_leaf': 3, 'min_samples_split': 8, 'n_estimators': 1000, random_state=0}

Gradient Boost: Best Parameters: {'learning_rate': 0.5, 'max_depth': 15 'max_features': 'sqrt', 'min_samples_split': 50, 'n_estimators': 80, random_state=10}

LightBGM: Best Parameters: {'learning_rate': 0.5, 'max_depth': 20, 'metric': 'binary_logloss', 'num_leaves': 20, 'objective': 'binary', random_state=10}

```python
from sklearn.linear_model import LogisticRegression #Logistic Regression
from sklearn.naive_bayes import GaussianNB #Naive Bayes
from sklearn.ensemble import GradientBoostingClassifier  #GBM algorithm
from sklearn.ensemble import RandomForestClassifier #Random Forest
import lightgbm as lgb #lightGBM

#Metrics Libraries
from sklearn import metrics
from sklearn.metrics import classification_report, plot_roc_curve
from sklearn.model_selection import cross_val_score
```

```python
# explore is miss.xxx features are important
# models: - logistic regression (ZJ); naïve baysian (Mansi); random forest (ZJ); gradient boost (ZJ); boosting-light GBM (Mansi); CNN (ZJ); BERT (Mansi);  

X = data.drop('fraudulent',1)
y = data.fraudulent


X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size = 0.30, random_state = 0)

print('Shape of X_train =', X_train.shape, ' and shape of y_train = ', y_train.shape)
print('Shape of X_test =', X_test.shape, ' and shape of y_test = ', y_test.shape)
```

```python
Logreg=LogisticRegression()
nbm = GaussianNB()
gdboost=GradientBoostingClassifier(learning_rate=0.5, max_depth=15, max_features='sqrt', min_samples_split=50, n_estimators=80, random_state=10)
rf=RandomForestClassifier(max_depth=50, max_features=4, min_samples_leaf=3, min_samples_split=8, n_estimators=1000, random_state=0)
lgbm=lgb.LGBMClassifier(learning_rate=0.5, max_depth=20, metric='binary_logloss', num_leaves=50, objective='binary', random_state=10)
```

```python
#Train models

Logreg.fit(X_train,y_train)
nbm.fit(X_train,y_train)
gdboost.fit(X_train,y_train)
rf.fit(X_train,y_train)
lgbm.fit(X_train,y_train)
print('Models are trained')
```

```python
Logreg_yprad = Logreg.predict(X_test)
nbm_yprad = nbm.predict(X_test)
rf_yprad = rf.predict(X_test)
gdboost_yprad = gdboost.predict(X_test)
lgbm_yprad = lgbm.predict(X_test)
```

```python
# AUC plot
classifiers = [Logreg,nbm,rf,gdboost,lgbm]
ax = plt.gca()
for i in classifiers:
    
    plot_roc_curve(i, X_test, y_test, ax=ax)
```

## **SMOTE-treated**

```python
from imblearn.over_sampling import SMOTE 
```

```python
# Use SMOTE to handle unbalanced data in train set

sm = SMOTE(random_state=42)

X_train_sm, y_train_sm = sm.fit_resample(X_train, y_train)

print(f'''Shape of X before SMOTE: {X_train.shape}
Shape of X after SMOTE: {X_train_sm.shape}''')

print('\nBalance of positive and negative classes (%):')
y_train_sm.value_counts(normalize=True) * 100
```

```python
print('\nBalance of positive and negative Revenue classes for training set (%):')
y_train_sm.value_counts(normalize=True) * 100
```

```python
Logreg_sm=LogisticRegression()
nbm_sm = GaussianNB()
gdboost_sm=GradientBoostingClassifier(learning_rate=0.5, max_depth=15, max_features='sqrt', min_samples_split=50, n_estimators=80, random_state=10)
rf_sm=RandomForestClassifier(max_depth=50, max_features=4, min_samples_leaf=3, min_samples_split=8, n_estimators=1000, random_state=0)
lgbm_sm=lgb.LGBMClassifier(learning_rate=0.5, max_depth=20, metric='binary_logloss', num_leaves=50, objective='binary', random_state=10)
```

```python
Logreg_sm.fit(X_train_sm,y_train_sm)
nbm_sm.fit(X_train_sm,y_train_sm)
gdboost_sm.fit(X_train_sm,y_train_sm)
rf_sm.fit(X_train_sm,y_train_sm)
lgbm_sm.fit(X_train_sm,y_train_sm)
print('SM Models are trained')
```

```python
Logreg_yprad_sm = Logreg_sm.predict(X_test)
nbm_yprad_sm = nbm_sm.predict(X_test)
rf_yprad_sm = rf_sm.predict(X_test)
gdboost_yprad_sm = gdboost_sm.predict(X_test)
lgbm_yprad_sm = lgbm_sm.predict(X_test)
```

```python
# AUC plot
classifiers = [Logreg_sm,nbm_sm,rf_sm,gdboost_sm,lgbm_sm]
ax = plt.gca()
for i in classifiers:
    
    plot_roc_curve(i, X_test, y_test, ax=ax)
```
